```python
teamlist = list()
for i in range(90):
    teamdata = mlb.lookup_team(i+70)
    if len(teamdata) > 0:
        teamlist.append(teamdata)
```

```python
team1 = list()
for item in teamlist:
    t = list()
    t.append(item[0]['id'])
    t.append(item[0]['name'])
    team1.append(t)
    
    
    
    
```

```python
df1 = pd.DataFrame(team1)
```

```python
df1.to_csv("teams_with_codes.csv")
```

```python
year = 2018
team25manRoster2018 = list()
for i in range(30):
    team_code = df1.iloc[i,0]
    team_name = df1.iloc[i,1]
    roster_a = mlb.roster(team_code,'25Man',date=mlb.get('season',{'seasonId':year,'sportId':1})['seasons'][0]['regularSeasonStartDate'])
    roster_b = roster_a.split('\n')
    roster_c = list(dict.fromkeys(roster_b))        
    for item in roster_c:
        player=list()
        player.append(team_code)
        player.append(team_name)
        roster_d = item.split(' ')
        for word in roster_d:
            if not word == '':
                player.append(word)        
        team25manRoster2018.append( player )                
                    
                    
```
